```python
# METACRITIC MOVIES ASSIGNMENT - JUPYTER NOTEBOOK
# CHIDOZIE OKOYE
# DATA 320 - T. SPENCE

# setting up access to websites

# import required modules for project

import urllib3
import certifi
import pandas as pd

# establish connection to the web

http = urllib3.PoolManager(ca_certs=certifi.where())

# access the website and set data tupe so we can read it with regex

link = "https://www.metacritic.com/browse/movie/all/all/2023/metascore/?page=1"

r = http.request('GET', link, headers={"User-Agent":"Chrome/86.0.4240.111"})

datastring = str(r.data, "utf-8")

# print a status for the established connection

print("CONNECTION STATUS:\n")
print(f"The request status is {r.status}")
print(f"The request from {link} fetched {len(datastring)} bytes")
print("\n")

# now time to retrieve data from my website source

# import regex module

import re

# locate needed data from source

movieTitle = re.findall("<div data-title=\"(.*?)\"", datastring)
releaseDate = re.findall("u-text-uppercase\">\s+(.*)", datastring)
metascore = re.findall("Metascore (.*?) out of 100\" aria", datastring)
imageUrl = re.findall("c-cmsImage\"> <img src=\"(.*?)\"", datastring)
description = re.findall("description\"><span>([\\s\\S]*?)<\\/span><\\/div>", datastring)

# correct "&amp;" issue in image urls

fixedUrl = []
for url in imageUrl:
    url = url.replace("&amp;", "&")
    fixedUrl.append(url)

# print confirmation of retrieved data

print("PREVIEW OF RETRIEVED DATA:\n")

print(f"Found {len(movieTitle)} movies: {movieTitle}")
print(f"Found {len(releaseDate)} dates: {releaseDate}")
print(f"Found {len(metascore)} scores: {metascore}")
print(f"Found {len(fixedUrl)} urls: {fixedUrl}")
print(f"Found {len(description)} descriptions: {description}")
```

```python
# we need a dataframe for our data so lets create one

movies = pd.DataFrame(
    {'Movie Title': movieTitle,
    'Release Date': releaseDate,
    'Metascore': metascore,
    'Image URL': fixedUrl,
    'Description': description
    })

# set display for optimal view

pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# nice printout of data

print(movies)
```
